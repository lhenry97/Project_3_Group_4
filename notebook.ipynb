```python
import requests
import json
from config import api_key
import pandas as pd
```

```python
url = "https://www.carboninterface.com/api/v1/estimates"
headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json"
}
```

```python
def get_emissions_estimate_country(electricity_value, country):
    payload = {
        "type": "electricity",
        "electricity_unit": "kwh",
        "electricity_value": electricity_value,
        "country": country
    }
    response = requests.post(url, headers=headers, json=payload)
    return response.json()

# Example usage
countries = ["US","CA","AT","BE","BG","HR","CY","CZ","DK","EU-27","EU-27+1","EE","FI","FR","DE","GR","GU","IE","IT","LV","LT","LU","MT","NL","PL","PT","RO","SK","SI","ES","SE","GB"]

for country in countries:
    estimate = get_emissions_estimate_country(100, country)
    print(json.dumps(estimate, indent=4))
```

```python
def get_emissions_estimate_state(electricity_value, country, state):
    payload = {
        "type": "electricity",
        "electricity_unit": "kwh",
        "electricity_value": electricity_value,
        "country": country,
        "state": state
    }
    response = requests.post(url, headers=headers, json=payload)
    return response.json()

canada_states = ["AB","BC","MB","NB","NL","NT","NS","NU","ON","PE","QC","SK","YT"]
for state in canada_states:
    estimate = get_emissions_estimate_state(100, "ca", state)
    print(json.dumps(estimate, indent=4))
```

```python
us_states = [ "AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY", "DC", "AS", "GU", "MP", "PR", "UM", "VI" ]
for state in us_states:
    estimate = get_emissions_estimate_state(100, "us", state)
    print(json.dumps(estimate, indent=4))
```

```python
# Estimate request
def get_emissions_estimate(activity_type, params):
    payload = {"type": activity_type}
    payload.update(params)
    response = requests.post(url, headers=headers, json=payload)
    return response.json()

activities = [
    {"type": "electricity", "params": {"electricity_unit": "kwh", "electricity_value": 100, "country": "us", "state": "ca"}},
    {"type": "flight", "params": {"passengers": 1, "legs": [{"departure_airport": "SFO", "destination_airport": "LAX"}]}},
    {"type": "vehicle", "params": {"distance_unit": "mi", "distance_value": 100, "vehicle_model_id": "some_vehicle_model_id"}},
    # Add more activities as needed
]

results = []
for activity in activities:
    activity_type = activity["type"]
    params = activity["params"]
    estimate = get_emissions_estimate(activity_type, params)
    results.append(estimate)
print(results)
```
