```python
from geopy.geocoders import Nominatim

# Initialize Nominatim API
geolocator = Nominatim(user_agent="geoapiExercises")

# Function to get the longitude and latitude of a country
def get_coordinates(country_name):
    location = geolocator.geocode(country_name)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Apply the function to your DataFrame to get the coordinates for each country
country_grouped['latitude'], country_grouped['longitude'] = zip(*country_grouped['country'].apply(get_coordinates))

# Display the updated DataFrame with latitude and longitude
country_grouped.head()
```
