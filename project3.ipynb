```python
# Define the API URL and headers
url = "https://www.carboninterface.com/api/v1/estimates"
headers = {
    "Authorization": f"Bearer {'7rme1XCm0LsZlEt5z4FaPA'}",
    "Content-Type": "application/json"
}
def get_emissions_estimate_country(electricity_value, country):
    try:
        payload = {
            "type": "electricity",
            "electricity_unit": "kwh",
            "electricity_value": electricity_value,
            "country": country
        }
        response = requests.post(url, headers=headers, json=payload)
        data = response.json()
        # Extract only the needed attributes
        attributes = data.get('data', {}).get('attributes', {})
        selected_data = {
            "country": attributes.get("country"),
            "state": attributes.get("state"),
            "electricity_unit": attributes.get("electricity_unit"),
            "electricity_value": attributes.get("electricity_value"),
            "estimated_at": attributes.get("estimated_at"),
            "carbon_g": attributes.get("carbon_g"),
            "carbon_lb": attributes.get("carbon_lb"),
            "carbon_kg": attributes.get("carbon_kg"),
            "carbon_mt": attributes.get("carbon_mt")
        }
        return selected_data
    except Exception as e:
        # Optionally log the error
        # print(f"An error occurred for {country}: {e}")
        return None

# Example usage
countries = ["US", "CA", "AT", "BE", "BG", "HR", "CY", "CZ", "DK", "EU-27", "EU-27+1", "EE", "FI", "FR", "DE", "GR", "GU", "IE", "IT", "LV", "LT", "LU", "MT", "NL", "PL", "PT", "RO", "SK", "SI", "ES", "SE", "GB"]

results = []
for country in countries:
    estimate = get_emissions_estimate_country(100, country)
    if estimate:
        results.append(estimate)

# Convert the list of results to a DataFrame
df = pd.DataFrame(results)

# Display the DataFrame
```
